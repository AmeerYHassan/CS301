#### Question 4 - Using the original dataframe (unchanged from Questoins 2, 3), create a mapping for the "Punxsutawney Phil" column which changes the categories into numerical quantities. For example: 'No Record' can be 0, 'No Shadow' can be 1, 'Partial Shadow' can be 2, and 'Full Shadow' can be 3. You should create the mapping *and* apply it to the data set by adding a new column called "Shadow".

```python
mappingDict = {'No Record':0, 'No Shadow':1, 'Partial Shadow':2, 'Full Shadow':3}
df["Shadow"] = df["Punxsutawney Phil"].map(mappingDict)
df
```

#### Question 5 - For all rows which have no record, remove the rows which have empty values in all of the quantitative columns. (Only 6 should be removed.)

```python
for index, row in df.iterrows():
    if row['Shadow'] is 0 and math.isnan(row['February Average Temperature']):
        print (f"Removing {row['Year']}")
        df = df.drop(index)
    elif row['Shadow'] is 0:
        print (f"{row['Year']} has no shadow")
df.reset_index(drop=True, inplace=True)
df
```

#### Question 6 - For the remaining rows which have no record (but the other columns *are* populated), determine an appropriate method for inputting the value for this column. Explain your reasoning, and show your work below (using pandas).

```python
fullShadowAvg = [0, 0, 0]
partialShadowAvg = [0, 0, 0]
noShadowAvg = [0, 0, 0]

for index, row in df.iterrows():
    if(not math.isnan(row["February Average Temperature"]) and not math.isnan(row["March Average Temperature"])):
        if row["Shadow"] is 3:
            fullShadowAvg[0]+=1
            fullShadowAvg[1]+=row["February Average Temperature"]
            fullShadowAvg[2]+=row["March Average Temperature"]
        if row["Shadow"] is 2:
            partialShadowAvg[0]+=1
            partialShadowAvg[1]+=row["February Average Temperature"]
            partialShadowAvg[2]+=row["March Average Temperature"]
        if row["Shadow"] is 1:
            noShadowAvg[0]+=1
            noShadowAvg[1]+=row["February Average Temperature"]
            noShadowAvg[2]+=row["March Average Temperature"]

fullShadowAvg[1] = fullShadowAvg[1]/fullShadowAvg[0]
fullShadowAvg[2] = fullShadowAvg[2]/fullShadowAvg[0]
partialShadowAvg[1] = partialShadowAvg[1]/partialShadowAvg[0]
partialShadowAvg[2] = partialShadowAvg[2]/partialShadowAvg[0]
noShadowAvg[1] = noShadowAvg[1]/noShadowAvg[0]
noShadowAvg[2] = noShadowAvg[2]/noShadowAvg[0]

print(f"Full Shadow Average Temps:\n\tFebruary: {fullShadowAvg[1]}\n\tMarch: {fullShadowAvg[2]}\n")
print(f"Partial Shadow Average Temps:\n\tFebruary: {partialShadowAvg[1]}\n\tMarch: {partialShadowAvg[2]}\n")
print(f"No Shadow Average Temps:\n\tFebruary: {noShadowAvg[1]}\n\tMarch: {noShadowAvg[2]}\n")
```
