# HW5 - Data Visualization

Importing necessary libraries to do this assignment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Data Reader method from previous homework to read the data.

```python
def data_reader(data_path):
    try:
        df = pd.read_csv(data_path)
        print(df.shape)
        return df
    except FileNotFoundError:
        print("Error during reading, please try again.")
        return None
```

## Problem 1: Bar Chart

Use a `bar chart` to display the median of the Midterm, Final, and Final Average columns, per major. Identify the data items, attributes, and the marks and channels used. Comment on the overall performance of each major...Which major(s) appear to be doing the best? Why do you think so?

**ANSWERS**

1. Data Items
    * The Data Items here are the median midterm, final, and final average grades.
2. Attributes
    * The score is the attribute, the major is being plotted against the score.
3. Marks
    * The marks here are the bars.
4. Channel
    * The channels here are the vertical position of the bars.

**Performance Analysis:**
    The BNFO major seems to be doing the best overall as they have the highest Final Average.

```python
studentDf = data_reader("student_class.csv")
```

```python
majorsList = ['BNFO', 'CS', 'IS', 'IT', 'MATH']
categoryList = ['Midterm', 'Final', 'Final Average']

medianDF = pd.DataFrame({'Majors':majorsList})

for major in majorsList:
    for category in categoryList:
        medianDF.loc[medianDF['Majors'] == major, category] = (studentDf.loc[studentDf.Majors == major]).loc[:,category].median()

w = 0.25
label_locations = np.arange(medianDF.shape[0])

plt.bar(label_locations, medianDF['Midterm'], width=w, label='Midterm', color='red')
plt.bar(label_locations + w, medianDF['Final'], width=w, label='Final', color='green')
plt.bar(label_locations + 2*w, medianDF['Final Average'], width=w, label='Final Average', color='blue')
plt.gcf().set_size_inches(14, 8)
plt.title('Median Midterm, Final, and Final Average scores')
plt.xlabel('Majors')
plt.ylabel('Score')
plt.xticks(label_locations+w, labels=majorsList)
plt.legend()
plt.show()
```

## Problem 2: Histogram

Use a `histogram` to plot the number of students in each grade bin. [40, 50), [50, 60),...etc. Identify the data items, attributes, and the marks and channels used. Critique the chart below. Is there anything you would change? Is there anything you would improve? Also comment on the distribution (approximately normal, skewed, etc). It should look like:


**ANSWERS**

1. Data Items
    * The Data Items here is the grade bins.
2. Attributes
    * The frequency is the attribute, the grade is being plotted against the frequency.
3. Marks
    * The marks here are the bars.
4. Channel
    * The channel here is the vertical position of the bar.

**Chart Critique:** An outline could be added to each of the rectangles so they don't necessarily all blend in with each other.
**Distribution:** This graph is left-skewed

```python
bins = list(set(studentDf['Numeric Bin']))
bins.append(100.0)
bins.sort()

plt.hist(x=studentDf['Final Average'], bins=bins, facecolor='blue', alpha=0.5)
plt.gcf().set_size_inches(12, 6)
plt.title('Number of students in each grade bin')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.show()
```
